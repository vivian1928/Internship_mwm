```python
import re
March_re = "(3\.\d{1,2}|3月\d{1,2}|3,\d{1,2})" # March
April_re = "(4\.\d{1,2}|4月\d{1,2}|4,\d{1,2})" # April
July_re = "(7\.\d{1,2}|7月\d{1,2}|7,\d{1,2})" # July
ESep_re = "(9.\d{1}-|9.\d{1}日|9.1[0-5]-|9.1[0-5]日)"# Early Sepetember
LSep_re = "(9.2\d{1}-|9.2\d{1}日|9.1[6-9]-|9.1[6-9]日|9.30-|9.30日)" # Late Sepetember
```

```python
March_row = []
April_row = []
July_row = []
ESep_row = []
LSep_row = []
No_row = []
for i in range(len(dataframe['date'])):
#     print(dataframe['date'][i])
    if type(dataframe['date'][i]) != str:
        No_row.append(dataframe.loc[i].values)
    else:
        if re.search(March_re,dataframe['date'][i]) != None:
            March_row.append(dataframe.loc[i].values)
        elif re.search(April_re,dataframe['date'][i]) != None:
            April_row.append(dataframe.loc[i].values)
        elif re.search(July_re,dataframe['date'][i]) != None:
            July_row.append(dataframe.loc[i].values)
        elif re.search(ESep_re,dataframe['date'][i]) != None:
            ESep_row.append(dataframe.loc[i].values)
        elif re.search(LSep_re,dataframe['date'][i]) != None:
            LSep_row.append(dataframe.loc[i].values)
        else:
            No_row.append(dataframe.loc[i].values)
            print('error:',end='')
            print(dataframe['date'][i])
```

```python
import operator

def sort_update_date(row,row_re,head):
    char1 = '年'
    char2 = '至'
    char3 = '-'
    date_re = '\d+'
    row_dict = {}
    index = 0
    end = '日'
    tmp_row = []
    for item in row:
        first_date = re.search(row_re, item[1])
        Pos1 = item[1].rfind(char1)
        Pos2 = item[1].rfind(char2)
        Pos3 = item[1].rfind(char3)
        Pos = max(Pos1, Pos2, Pos3)
        second_date = item[1][Pos+1:]
        first_date_all = re.findall(date_re, first_date.groups()[0])
        row_dict.update({index:int(first_date_all[1])})
        index += 1
    row_dictSort = sorted(row_dict.items(),key=lambda x:x[1],reverse=False)
    for item in row_dictSort:
        Pos1 = row[item[0]][1].rfind(char1)
        Pos2 = row[item[0]][1].rfind(char2)
        Pos3 = row[item[0]][1].rfind(char3)
        Pos = max(Pos1, Pos2, Pos3)
        second_date = row[item[0]][1][Pos+1:]
        first_date_all = re.findall(date_re, second_date)
        real_date = head+str(item[1])+end+char3+head+str(first_date_all[-1])+end
        row[item[0]][1] = real_date
        tmp_row.append(row[item[0]])
    return tmp_row
if __name__ == '__main__':
#     pdb.set_trace()
    new_row = []
    new_March_row = sort_update_date(March_row,March_re,'2021年3月')
    new_April_row = sort_update_date(April_row,April_re,'2021年4月')
    new_July_row = sort_update_date(July_row,July_re,'2021年7月')
    new_ESep_row = sort_update_date(ESep_row,ESep_re,'2021年9月')
    new_LSep_row = sort_update_date(LSep_row,LSep_re,'2021年9月')
    for item in new_March_row:
        new_row.append(item)
    for item in new_April_row:
        new_row.append(item)
    for item in new_July_row:
        new_row.append(item)
    for item in new_ESep_row:
        new_row.append(item)
    for item in new_LSep_row:
        new_row.append(item)
    for item in No_row:
        new_row.append(item)
    new_df=pd.DataFrame(new_row,columns=colname)
    new_df.to_csv('sort.csv',index=False)
```
